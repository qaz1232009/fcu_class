<font face="黑体" size=5>**簡單圖片處理**

<font face="黑体" size=4 >導入套件

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

<font face="黑体" size=4 >讀取圖片

```python
img = cv2.imread('image.jpg')
```

<font face="黑体" size=4 >查看資料型態

```python
print(type(img))
print(img.shape)
```

<font face="黑体" size=4 >NumPy 陣列的大小

<font face="黑体" size=4 >讀入的照片用matplotlib顯示出來
    

```python
plt.imshow(img)
```

<font face="黑体" size=4 >
  將 BGR 圖片轉為 RGB 圖片（將numpy陣列左右翻轉）

```python
img_rgb = img[:,:,::-1]
plt.imshow(img_rgb)
```

<font face="黑体" size=3 >這張圖片是257*196 彩色圖片<>(3代表RGB三色)

<font face="黑体" size=4 >轉成灰階 

```python
image_g = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
image_g.shape
```

```python
plt.imshow(image_g, cmap = 'gray')
```

<font face="黑体" size=4 >照片resize

```python
import imutils
img = cv2.imread('human.jpg')
```

```python
print(type(img))
print(img.shape)
```

```python
img_rgb = img[:,:,::-1]
plt.imshow(img_rgb)
```

```python
# 縮小圖片
img = imutils.resize(img, width=320)
print(type(img))
print(img.shape)
```

```python
img_rgb = img[:,:,::-1]
plt.imshow(img_rgb)
```

<font face="黑体" size=4 >儲存

```python
cv2.imwrite("save.jpg", img)
```

```python
img = cv2.imread('save.jpg')
print(type(img))
print(img.shape)
img_rgb = img[:,:,::-1]
plt.imshow(img_rgb)
```
